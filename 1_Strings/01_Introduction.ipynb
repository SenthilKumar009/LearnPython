```python
# Length

# To find the length of the string we should use len() function

text = "Data Engineer"
print(f"Length of the given String: {len(text)}")
```

```python
# Access index values

print(text[0]) # To print the first char of the string
print(text[-1]) # To print the last char of the string
```

```python
# Access index values using range (slicing)

print(text[0:2]) # Speficy the start and end range
print(text[1:]) # specifify the 
print(text[:4]) # Start from the 0 and end with specific end
```

```python
n = 10
print("Total Values is:", n)

x = "There are %d types of people." % 10 
# To use the integer values inside the sting notation we should use %d and specify the value with prefix % in the end

binary = "binary"
do_not = "don't"
y = "Those who know %s and those who %s." % (binary, do_not)

print(x)
print (y)

print("I said: %r." % x)
print("I also said: '%s'." % y)

hilarious = False
joke_evaluation = "Isn't that joke so funny?! %r"

print(joke_evaluation % hilarious)

w = "This is the left side of..."
e = "a string with a right side."

print(w + e)
```

```python
# String Methods:

text = 'Data Engineer'
print("Lower Case: ",text.lower())
print("Upper Case: ",text.upper())
print("Count word in the String: ",text.count('Data'))
print("Count letter in the String: ",text.count('a'))

# Find method
# It will retun the starting index of the word to be find
# It will return the -1 index value if the give word is not found in the sentance
print("Find word in the String: ",text.find('Data')) 

# Replace method
text.replace("Data", "Big Data") 
print(text)
# This wont work, since the given line creates a new sting and we need to define the variable to store the output
text_update= text.replace("Data", "Big Data") 
print(text_update)
```

```python
# Concate String(s)

greeting = "Hello"
name ="SKK"

print(greeting + ", " + name) # use + operator

print('{}, {}. Welcome!'.format(greeting, name)) # use format string

print(f'{greeting}, {name}. Welcome!') # use f string
```
